```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters (choose realistic demand scale)
mu = 1_200_000      # mean demand (1.2 million units)
sigma = 250_000     # standard deviation

# X range for plotting
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Gaussian density
y = norm.pdf(x, mu, sigma)

# 80% confidence interval bounds (10th and 90th percentiles)
lower = norm.ppf(0.10, mu, sigma)
upper = norm.ppf(0.90, mu, sigma)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, linewidth=2)

# Shade 80% interval
mask = (x >= lower) & (x <= upper)
plt.fill_between(x[mask], y[mask], alpha=0.3)

# Vertical lines for bounds and mean
plt.axvline(lower, linestyle="--", linewidth=2, label="10th percentile")
plt.axvline(upper, linestyle="--", linewidth=2, label="90th percentile")
plt.axvline(mu, linestyle="-", linewidth=2, label="Expected demand")

# Labels
plt.xlabel("Demand (units)")
plt.ylabel("Probability density")
plt.title("Distributional Forecast with 80% Confidence Interval")

# Text annotation
plt.text(mu, max(y)*0.9, "Expected demand", ha="center")
plt.text(lower, max(y)*0.6, "80% interval", ha="right")
plt.text(upper, max(y)*0.6, "80% interval", ha="left")

plt.legend()
plt.grid(True)
plt.show()
```
